## Including only "Instruments" having only either "FUTIDX" or "FUTSTX" values in the output

```python
from pyspark.sql.functions import col

curr_data= curr_data.filter(col("INSTRUMENT").isin(["FUTIDX","FUTSTK"]))
prev_data = prev_data.filter(col("INSTRUMENT").isin(["FUTIDX","FUTSTK"]))
```

## Now Filtering the index on the nearest Expiry Date after coverting date column into dd-mm-yyy format

```python
from pyspark.sql.functions import to_date,regexp_replace
import pandas as pd

today = pd.Timestamp.now()
last_day = today.replace(day=1) + pd.offsets.MonthEnd()

last_thursday = last_day
while last_thursday.weekday() !=3:
    last_thursday -=pd.Timedelta(days=1)
    
expiry_date =last_thursday.strftime("%d-%b-%Y")

print(expiry_date, type(expiry_date) )
```

## Filtering the data based on the expiry date

```python
curr_data = curr_data.filter(col("EXPIRY_DT") == expiry_date )
prev_data = prev_data.filter(col("EXPIRY_DT") == expiry_date )
```

```python
print(curr_data.count())
print(prev_data.count())
```

## joining the two csv files

## Calculating the change in Open Interset and  change in closing price

```python
new_master = master_df.withColumn('CHANGE_IN_LTP', ((col('CLOSE')-col('prev_close'))/col('prev_close'))*100 )
```

```python
new_master.printSchema()
```

```python
new_master = new_master.withColumn('CHANGE_IN_OI', ((col('OPEN_INT')-col('PREV_OPEN_INT'))/col('PREV_OPEN_INT'))*100 )
```

```python
new_master.printSchema()
```

```python
new_master.describe("CHANGE_IN_LTP").show()
new_master.describe("CHANGE_IN_OI").show()
```

```python
new_master.count()
```

**Visualizing the distribution of CHANGE_IN_LTP values**
** Higher the distribution towards change in extreme ends higher the chance of considering the stock**

```python
from pyspark.ml.feature import Bucketizer
from pyspark.sql.functions import col

# create a bucketizer to bin the values of CHANGE_IN_LTP column
bucketizer = Bucketizer(splits=[float("-inf"), -10, -5, 0, 5, 10, float("inf")], inputCol="CHANGE_IN_LTP", outputCol="bin")

# apply the bucketizer to the DataFrame
binned_data = bucketizer.transform(new_master)

# use the groupBy and count functions to count the number of rows in each bin
histogram_data = binned_data.groupBy("bin").count().orderBy("bin")

# display the histogram
histogram_data.show()
```

## Filter for LONG BUILD UP 10% or above OI and 2% or above close change

```python
long_buildup=new_master.filter((col('CHANGE_IN_OI') > 8.5) & (col('CHANGE_IN_LTP') > 2) ).select(col('INSTRUMENT'),col('SYMBOL'),col('CHANGE_IN_OI'),col('CHANGE_IN_LTP'))
```

```python
from pyspark.sql.functions import lit

longbuildupdf=long_buildup.withColumn("CALL",lit("Buy CE or Sell PE if uptrend confirms")).drop("LONG_BUILDUP")
```

```python
longbuildupdf=longbuildupdf.withColumn("CALL_TYPE",lit("LONG_BUILDUP"))
```
